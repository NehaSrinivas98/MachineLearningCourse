1. Write a program that prints a 5x5 grid of asterisks (*) using nested loops.

```python
for i in range(5):
    for j in range(5):
        print("*", end=" ")
    print()
```

2. Write a program that asks the user to input numbers until they input 0. The program should print the sum of all the input numbers.

```python
total = 0
while True:
    res = int(input("Input a number: "))
    if res == 0:
        break
    total += res

print(total)
```

3.Write a program that prints all the numbers from 1 to 10 except 5 using a for loop and continue statement.

```python
for i in range(1,11):
  if i == 5:
    continue
  print(i)
```

4.  Write a program that asks the user to input a number and prints all the even numbers from 1 to that number using a for loop.

```python
input_number = int(input("Input a number: "))
for i in range(1,input_number+1):
  if i%2 != 0:
    continue
  print(i)
```

5.Write a program that calculates the factorial of a number input by the user using a while loop.

```python
number = int(input("Enter a number: "))
factorial = 1
i = 1
while i <= number:
    factorial *= i
    i += 1
print(f"The factorial of {number} is {factorial}.")
```

6.Write a program that calculates the sum of the digits of a number input by the user using a while loop.

```python
number = int(input("Enter a number: "))
sum_of_digits = 0
while number > 0:
    digit = number % 10
    sum_of_digits += digit
    number = number // 10
print(f"The sum of the digits is {sum_of_digits}.")
```

7. Prime number check

```python
number = int(input("Enter a number: "))
is_prime = True
if number <= 1:
    is_prime = False
else:
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            is_prime = False
            break
if is_prime:
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

8.Write a program that prints the first n Fibonacci numbers, where n is input by the user.

```python
n = int(input("Enter the number of Fibonacci numbers to print: "))
a, b = 0, 1
count = 0
while count < n:
    print(a)
    a, b = b, a + b
    count += 1
```
