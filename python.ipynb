## **c. Set**

Sets are a built-in data type in Python used to store collections of unique items. They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference, and symmetric difference.

```python
set_variable = {10,'string',True,[],10.2}
# here error is thrown because of the empty list
```

```python
set_example_1 = {1,2,3,4.5,6,'test',True}
```


1.   Sets are immutable.
2.   Sets cannot have same value.

```python
set_example_2 = {1,2,3,4,1,2}
print(set_example_2)
```

```python
# Remove multiple values from the list
list_example = [1,2,3,5,6,1,3,4,5]
unique_values = list(set(list_example))
print(unique_values)
```

```python
set_example_2.add(7)
print("add", set_example_2)
set_example_2.remove(3)
print("remove", set_example_2)
# discard doesnt throw any error if the element doesnt exist
set_example_2.discard(11)
print("discard", set_example_2)
removed_element=set_example_2.pop()
print("removed element", removed_element)
print("after removed element",set_example_2)
```

```python
#clear all the elemets of the set
set_example_2.clear()
print("clear",set_example_2)
```

```python
## Set Memebership test
my_set={1,2,3,4,5}
print("3 is present in test", 3 in my_set)
print("10 is present in test", 10 in my_set)
```

```python
## MAthematical Operation
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

### Union
union_set=set1.union(set2)
print("union",union_set)

## Intersection
intersection_set=set1.intersection(set2)
print("intersection",intersection_set)

set1.intersection_update(set2)
print("intersection update",set1)

set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

## Difference
print("difference",set1.difference(set2))

## Symmetric Difference
print("symmetric difference",set1.symmetric_difference(set2))

## Sets Methods
set1={1,2,3,4,5}
set2={3,4,5}

## is subset
print("subset", set1.issubset(set2))

print("superset", set1.issuperset(set2))
```

```python
# Read
print("dict_exammple",dict_example)
print("access f using get", dict_example.get('f'))
print("access d using get else print not available", dict_example.get('d',"Not Available"))
print("access g using get else print not available", dict_example.get('g',"Not Available"))
print("access e element", dict_example['e'])
print("print all keys",dict_example.keys())
print("print all values", dict_example.values())
print("print all items", dict_example.items())
```
