# **2. Variables**

Variables are fundamental elements in programming used to store data that can be referenced and manipulated in a program. In Python, variables are created when you assign a value to them, and they do not need explicit declaration to reserve memory space. The declaration happens automatically when you assign a value to a variable.Python is dynamically typed,type of a variable is determined at runtime

```python
test = 'test'
```

***Rules for writing a variable***

1.   Variable names can only contain letters, digits and underscores (_).
2.   A variable name cannot start with a digit.
3.   Variable names are case-sensitive (myVar and myvar are different).
4.   Avoid using Python keywords (e.g., if, else, for) as variable names.

```python
tuple_variable = (10,'string',True,[],10.2)
print(tuple_variable)
```

```python
tuple_variable[0] = 20
```

if you want to modify a tuple convert it to list change the values and convert it back to a tuple

```python
tuple_example = (1,2,3,4,5)
converted_to_list = list(tuple_example)
converted_to_list[0] = 6
converted_to_tuple = tuple(converted_to_list)
print( converted_to_tuple )
```

## **d. Dictionary**
 they are mutable

```python
dict_variable = {'a':'1','b':2,'c':3.0,'d':[1,2],'e':(1,2),'f':{1,2} }
dict_variable['a']
```

```python
#Create
dict_example = {'a':'1','b':2,'c':3.0,'d':[1,2],'e':(1,2),'f':{1,2} }
```

```python
# Read
print(dict_example)
print(dict_example['e'])
print(dict_example.keys())
print(dict_example.values())
```

```python
#Update
dict_example['a'] = '2'
print(dict_example)
dict_example['g'] = {'0':'1'}
print(dict_example)
dict_example.update({'h':2})
print(dict_example)
```

```python
#Delete
print(dict_example['d'])
del dict_example["d"]
print(dict_example)
dict_example.pop("e")
print(dict_example)
dict_example.clear()
print(dict_example)
```

```python
#copy a dictionary
new_dict = dict_example.copy()
print(new_dict)
new_dict['a'] = 3
print(new_dict)
print(dict_example)
new_dict_1 = dict_example
new_dict_1['a'] = 4
print(new_dict_1)
print(dict_example)
```
